```python
import numpy as np
```

```python
def generate_random_points(geometry, num_points):
    min_x, min_y, max_x, max_y = geometry.bounds
    points_x = []
    points_y = []
    i=0
    while i<num_points:
        point_x = np.random.uniform(min_x, max_x)
        point_y = np.random.uniform(min_y, max_y)
        point = Point(point_x, point_y)
        if geometry.contains(point):
            points_x.append(point_x)
            points_y.append(point_y)
            i+=1
    return points_x,points_y
```

```python
merged_data['lat']=None
merged_data['lon']=None
```

```python
grouped_data = merged_data.groupby('iso3')
for group_name, group_df in grouped_data:
    print(f"Group Name: {group_name}")
```

```python
appended_df = pd.DataFrame()

# Iterate through each group in the grouped DataFrame
for group_name, group_data in grouped_data:
    # Create dummy 'lat' and 'lon' columns
    group_data['lat'] = None
    group_data['lon'] = None
    num_points = len(group_data)
    group_geometry=group_data['geometry'].iloc[0]
    # Modify the 'lat' and 'lon' columns as needed
    group_data['lon'], group_data['lat'] = generate_random_points(group_geometry, num_points)
    
    # Append the modified group to the new DataFrame
    appended_df = pd.concat([appended_df,group_data])

# Reset the index of the appended DataFrame
appended_df.reset_index(drop=True, inplace=True)
```
